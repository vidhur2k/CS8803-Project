# Experiment Results for Top-down and Bottom-up

**NOTE:** This is simply a compilation of the experiment results that we obtained from `topdownexperiments.ipynb`, `bottomupexperiments.ipynb`, and `ofa.ipynb`. We couldn't get the top-down and bottom-up methods to run in the same files as the original OFA method, so we used separate notebooks for them all and compiled the results here.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Experiment 1: Running Time

```python
def plot_bar(vanilla, top_down, bottom_up, path):
    latencies = np.array([1, 2, 3, 4, 5, 10])
    x = np.arange(len(latencies))
    width = 0.25
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.bar(x-width, vanilla, width=width, label='Vanilla')
    ax.bar(x+width, bottom_up, width=width, label='Bottom-up')
    ax.bar(x, top_down, width=width, label='Top-down')
    ax.set_xticks(x)
    ax.set_xticklabels(latencies)
    plt.legend()
    plt.xlabel('#Latency Constraints')
    plt.ylabel('Running Time (seconds)')
    plt.savefig(path, bbox_inches='tight')
```

```python
latencies = np.array([1, 2, 3, 4, 5, 10])
x = np.arange(len(latencies))
vanilla = np.array([27.0648090839386, 50.516709089279175, 68.79455637931824, 96.6699378490448, 123.10927200317383, 247.85819578170776])
top_down = np.array([15.352033138275146, 21.316713094711304, 26.708261013031006, 29.721980810165405, 35.34088611602783, 55.60786724090576])
bottom_up = np.array([26.16634511947632, 31.591012001037598, 38.2353081703186, 43.80901288986206, 47.78053402900696, 64.29084706306458])
```
